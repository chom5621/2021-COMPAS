```python
 # 추세 분석
from fbprophet import Prophet

# 추세 분석 함수
def trend(dt):
   
    #칼럼명 변경 
    ## prophet을 사용하기 위해서는 다음과 같이 피처의 이름을 변경해야 합니다 : 'ds', 'y'
    dt.columns=['ds','y']
    # datetime type로
    dt['ds']=pd.to_datetime(dt['ds'],format='%Y%m')
    
    # 모델생성
    prophet = Prophet(seasonality_mode='multiplicative', 
                  yearly_seasonality=True,
                  weekly_seasonality=True, daily_seasonality=True,
                  changepoint_prior_scale=0.5)
    # 모델학습
    prophet.fit(dt)
    # 모델예측 ~24년 6월까지
    future_data = prophet.make_future_dataframe(periods=39, freq='M')
    forecast_data = prophet.predict(future_data)
    rlt=forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
    return rlt
```

### ◉ 전체자동차등록수 추세분석

```python
# 주차가능차량: 승용 전체 + 경/소형 승합차

def 자동차등록현황(자동차등록수):
    자동차등록수 = 자동차등록수.rename(columns={'Unnamed: 0':'emd_info', 'Unnamed: 2':'차종', 'Unnamed: 4':'계',
                                        '소계':'non_bs_all','경형':'non_bs_city','소형':'non_bs_compact','중형':'non_bs_midsize','대형':'non_bs_fullsize',
                                    '소계.1':'bs_all','경형.1':'bs_city','소형.1':'bs_compact','중형.1':'bs_midsize','대형.1':'bs_fullsize'})

    자동차등록수= 자동차등록수.drop([0,1,2,3,4], axis=0)
    자동차등록수= 자동차등록수.fillna(method='ffill')# nan값을 행정동명으로 채워줌.
    
    전체합계=자동차등록수[자동차등록수.차종=='소계']
    전체합계['소계']=전체합계.non_bs_all + 전체합계.bs_all
    전체합계.set_index('emd_info',inplace=True)
    
    자동차등록수=자동차등록수[~자동차등록수.차종.isin(['화물','특수','소계'])] # 승용, 승합만 살려놓기
    자동차등록수.reset_index(drop=True,inplace=True) #필수

    # 주차가능차량 구하기
    전체합계['주차가능차량']=np.nan

    for dong in 전체합계.index:
        주차가능차량=0
        for i in 자동차등록수[자동차등록수.emd_info==dong].index:
            if 자동차등록수.차종[i]=='승용':
                주차가능차량=주차가능차량+자동차등록수.계[i]# 승용일 경우,
            else: # 승합일 경우
                주차가능차량=주차가능차량+자동차등록수[['non_bs_city','non_bs_compact','bs_city','bs_compact']].iloc[i].sum()
        전체합계['주차가능차량'].loc[dong]=주차가능차량

    전체합계=전체합계[['소계','주차가능차량']]
    
    return 전체합계
```
