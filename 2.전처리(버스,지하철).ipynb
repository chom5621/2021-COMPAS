```python
# 파일 임포트
버스=pd.read_csv('7.부산시남구_버스정류소정보.csv')
지하철=pd.read_csv('/8.부산시남구_지하철역정보.csv')    
```

```python
# 격자별
def multipolygon_to_coordinates(x): 
    lon, lat = x[0].exterior.xy 
    return [[x, y] for x, y in zip(lon, lat)] 

def polygon_to_coordinates(x): 
    lon, lat = x.exterior.xy 
    return [[x, y] for x, y in zip(lon, lat)] 
```

```python
import pyproj
from fiona.crs import from_epsg

버스위경도= gpd.GeoDataFrame(버스[['lon','lat']], geometry=gpd.points_from_xy(버스.lon,버스.lat))
지하철위경도=gpd.GeoDataFrame(지하철[['lon','lat']], geometry=gpd.points_from_xy(지하철.lon,지하철.lat))

gdf_버스위경도 = gpd.GeoDataFrame(버스위경도,crs=from_epsg(4326))
gdf_지하철위경도 = gpd.GeoDataFrame(지하철위경도,crs=from_epsg(4326))
```

```python
# GetCompasData('SBJ_2107_003', '15', '15.부산시남구_격자(300X300).geojson')
격자= gpd.read_file('15.부산시남구_격자(300X300).geojson') # 파일 임포트
# geometry를 coordinate 형태로 적용
# 함수 polygon_to_coordinates
격자['coordinates'] = 격자['geometry'].apply(polygon_to_coordinates) #pydeck 을 위한 coordinate type
격자
```

```python
#격자별 버스&지하철
격자별_버스=gpd.sjoin(격자, gdf_버스위경도, how='left', op='intersects')
격자별_지하철=gpd.sjoin(격자, gdf_지하철위경도, how='left', op='intersects')
```

```python
격자['버스']=격자별_버스.groupby('id',as_index=False)['lon'].count()['lon']
격자['지하철']=격자별_지하철.groupby('id',as_index=False)['lon'].count()['lon']
```

```python
# 패키지 호출
from sklearn.preprocessing import MinMaxScaler

# 객체 생성 
scaler1 = MinMaxScaler()
scaler2 = MinMaxScaler()

x1=격자['버스'].to_numpy()
x2=격자['지하철'].to_numpy()

x1= x1.reshape(-1,1)
x2= x2.reshape(-1,1)

# 모델 fit
버스_정규화= scaler1.fit_transform(x1)
지하철_정규화= scaler2.fit_transform(x2)
```

```python
격자['정규화_버스']=버스_정규화
격자['정규화_지하철']=지하철_정규화
```

```python
# 7.8. 격자별_버스+지하철_정규화 로 파일 저장
격자.to_csv('7.8. 격자별_버스+지하철_정규화')
```
