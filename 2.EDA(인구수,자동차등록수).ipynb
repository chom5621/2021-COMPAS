```python
# 대형차량 제외, 자동차등록수 계산함수
def 자동차등록현황(자동차등록수):
    자동차등록수 = 자동차등록수.rename(columns={'Unnamed: 0':'emd_info', 'Unnamed: 2':'차종', 'Unnamed: 4':'계',
                                        '소계':'non_bs_all','경형':'non_bs_city','소형':'non_bs_compact','중형':'non_bs_midsize','대형':'non_bs_fullsize',
                                    '소계.1':'bs_all','경형.1':'bs_city','소형.1':'bs_compact','중형.1':'bs_midsize','대형.1':'bs_fullsize'})

    자동차등록수 = 자동차등록수.drop(['계','non_bs_all','non_bs_fullsize','bs_all','bs_fullsize'], axis=1)
    자동차등록수 = 자동차등록수.fillna(method='ffill')
    자동차등록수.drop([0,1,2,3,4], axis=0, inplace=True)
    lst=[]
    for x in 자동차등록수.차종.index:
        if 자동차등록수.차종[x] == '소계':
            lst.append(x)
    자동차등록수 = 자동차등록수.loc[lst]
    자동차등록수.drop('차종', axis=1, inplace=True)
    자동차등록수['소계'] = 자동차등록수.non_bs_city + 자동차등록수.non_bs_compact + 자동차등록수.non_bs_midsize + 자동차등록수.bs_city + 자동차등록수.bs_compact + 자동차등록수.bs_midsize
    자동차등록수 = 자동차등록수[['emd_info','소계']].set_index('emd_info')
    return 자동차등록수
```

### [3] 자동차등록비율

* $ 인구당자동차등록수 = \frac{자동차등록수}{운전인구} $

#### 운전인구 (운전가능인구)
* $ 운전인구 = 총인구 - 미성년인구 - 고령인구 $
* 제1종 보통, 제1종 소형, 제2종 보통 및 제2종 소형면허는 만 18세 이상인 사람만 취득할 수 있으므로 (「도로교통법」제82조제1항제1호 및 제6호) \
미성년인구(0~18세)는 운전인구에서 제외
* 운전면허증 반납 65세 이상 어르신에게는 교통사랑카드 발급 등 혜택이 주어지므로 (부산남구신문 > https://www.bsnamgu.go.kr/board/view.namgu?boardId=BBS_0000088&menuCd=DOM_000000114001000000&startPage=50&dataSid=86412) \
고령인구(65세 이상)도 운전인구에서 제외

```python
GetCompasData('SBJ_2107_003', '9', '9.부산시남구_동별_인구정보.csv')
인구 = pd.read_csv('9.부산시남구_동별_인구정보.csv')

# 총인구 계산
총인구 = 인구[(인구['age']=='총계')&(인구['stndrd']=='계')].drop(['age','stndrd'], axis=1).set_index('dong').transpose().drop('남구', axis=1)
총인구 = 총인구.transpose()
총인구.drop(['201601','201602','201603','201604','201605','201606','201607','201608','201609','201610','201611','201612',
           '201701','201702','201703','201704','201705','201706','201707','201708','201709','201710','201711','201712',
           '202104'], axis=1, inplace=True)
총인구['인구평균']=총인구.mean(axis=1)
총인구 = 총인구['인구평균']
총인구 = pd.DataFrame(총인구)


# 65세 이상 인구 계산
연령별인구 = 인구[인구['stndrd']=='계'].drop(['stndrd'], axis=1).set_index(['dong','age']).transpose().drop('남구', axis=1)

lst=[]
for x,y in 연령별인구.columns:
    if y=='65세 이상':
        lstt = 연령별인구[(x,y)]
        lst.append(lstt)
고령인구 = pd.DataFrame(lst)

고령인구.drop(['201601','201602','201603','201604','201605','201606','201607','201608','201609','201610','201611','201612',
           '201701','201702','201703','201704','201705','201706','201707','201708','201709','201710','201711','201712',
           '202104'], axis=1, inplace=True)
고령인구['고령인구평균'] = 고령인구.mean(axis=1)
고령인구 = 고령인구['고령인구평균']
고령인구 = pd.DataFrame(고령인구).sort_values(by='고령인구평균', ascending=False)
총인구 = pd.concat([총인구, 고령인구.rename(index={('용호1동', '65세 이상'):'용호1동', ('대연3동', '65세 이상'):'대연3동', ('대연1동', '65세 이상'):'대연1동',
        ('우암동', '65세 이상'):'우암동', ('감만1동', '65세 이상'):'감만1동', ('용호2동', '65세 이상'):'용호2동',
        ('용호3동', '65세 이상'):'용호3동', ('대연5동', '65세 이상'):'대연5동', ('문현1동', '65세 이상'):'문현1동',
        ('대연4동', '65세 이상'):'대연4동', ('문현3동', '65세 이상'):'문현3동', ('문현4동', '65세 이상'):'문현4동',
        ('대연6동', '65세 이상'):'대연6동', ('문현2동', '65세 이상'):'문현2동', ('감만2동', '65세 이상'):'감만2동',
        ('용호4동', '65세 이상'):'용호4동', ('용당동', '65세 이상'):'용당동'},columns = {0:'고령인구평균'})], axis=1)


# 미성년인구 계산
lst=[]
for x,y in 연령별인구.columns:
    if y=='0~14세' or y=='15세' or y=='16세' or y=='17세' or y=='18세':
        lstt = 연령별인구[(x,y)]
        lstt.name = lstt.name[0]
        lst.append(lstt)
미성년인구 = pd.DataFrame(lst)

미성년인구.drop(['201601','201602','201603','201604','201605','201606','201607','201608','201609','201610','201611','201612',
           '201701','201702','201703','201704','201705','201706','201707','201708','201709','201710','201711','201712',
           '202104'], axis=1, inplace=True)
미성년인구 = 미성년인구.groupby(미성년인구.index).sum()   

미성년인구['미성년인구평균'] = 미성년인구.mean(axis=1)
미성년인구 = 미성년인구['미성년인구평균']
미성년인구 = pd.DataFrame(미성년인구).sort_values(by='미성년인구평균', ascending=False)


# 운전인구 계산
총인구 = pd.merge(총인구, 미성년인구, left_index=True, right_index=True, how='left')
총인구['운전인구평균'] = 총인구['인구평균'] - 총인구['미성년인구평균'] - 총인구['고령인구평균']
총인구.drop(['고령인구평균','미성년인구평균'],axis=1,inplace=True)
총인구.sort_values(by='운전인구평균', ascending=False)
```

#### 격자별 자동차등록비율 시각화를 위한 기초작업 

#### 1. 격자별 중심점(centroid) 뽑기 

```python
GetCompasData('SBJ_2107_003', '15', '15.부산시남구_격자(300X300).geojson')
df= gpd.read_file("15.부산시남구_격자(300X300).geojson")

# 격자 시각화를 위한 함수 호출
def multipolygon_to_coordinates(x): 
    lon, lat = x[0].exterior.xy 
    return [[x, y] for x, y in zip(lon, lat)] 
def polygon_to_coordinates(x): 
    lon, lat = x.exterior.xy 
    return [[x, y] for x, y in zip(lon, lat)] 

# geometry를 coordinate 형태로 바꾸기
df['coordinates'] = df['geometry'].apply(polygon_to_coordinates)

# 300X300 grid에서 central point 찾기
df_list = []
for i in df['geometry']:
    cent = i.centroid
    df_list.append(cent)
df['geo_cent'] = 0
df['geo_cent'] = df_list 

# 쉬운 분석을 위한 임의의 grid id 생성
df['grid_id']=0
idx = []
for i in range(len(df)):
    idx.append(str(i).zfill(5))
df['grid_id'] = pd.DataFrame(idx)

df.head(2)
```

```python
# 대형차량 제외, 승용차등록비율 계산함수
def 승용차비율(자동차등록수):
    자동차등록수 = 자동차등록수.rename(columns={'Unnamed: 0':'emd_info', 'Unnamed: 2':'차종', 'Unnamed: 4':'계',
                                        '소계':'non_bs_all','경형':'non_bs_city','소형':'non_bs_compact','중형':'non_bs_midsize','대형':'non_bs_fullsize',
                                    '소계.1':'bs_all','경형.1':'bs_city','소형.1':'bs_compact','중형.1':'bs_midsize','대형.1':'bs_fullsize'})

    자동차등록수 = 자동차등록수.drop(['계','non_bs_all','non_bs_fullsize','bs_all','bs_fullsize'], axis=1)
    자동차등록수 = 자동차등록수.fillna(method='ffill')
    자동차등록수.drop([0,1,2,3,4], axis=0, inplace=True)
    lst=[]
    for x in 자동차등록수.차종.index:
        if 자동차등록수.차종[x] == '승용':
            lst.append(x)
    자동차등록수 = 자동차등록수.loc[lst]
    자동차등록수.drop('차종', axis=1, inplace=True)
    자동차등록수['승용'] = 자동차등록수.non_bs_city + 자동차등록수.non_bs_compact + 자동차등록수.non_bs_midsize + 자동차등록수.bs_city + 자동차등록수.bs_compact + 자동차등록수.bs_midsize
    자동차등록수 = 자동차등록수[['emd_info','승용']].set_index('emd_info')
    return 자동차등록수
```
